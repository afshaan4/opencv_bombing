```python
# get the data
%matplotlib inline
from pylab import *
import numpy as np
from numpy import genfromtxt

data = genfromtxt('scaling2.csv', delimiter = ',')

data = np.delete(data, (0), axis = 0) # remove the header
# get separate the columns
distances = data[:, [0]]
xLength = data[:, [1]]
yLength = data[:, [2]]
# plot it
scatter(distances, xLength)
scatter(distances, yLength)
```

```python
# now try and fit a polynomial function to it

x = xLength[:, 0]
y = yLength[:, 0]
d = distances[:, 0]

polynomial1 = np.poly1d(np.polyfit(d, x, 8)) # x, y, degree
polynomial2 = np.poly1d(np.polyfit(d, y, 8)) # x, y, degree
```

```python
import matplotlib.pyplot as plt

xp = np.linspace(30, 350, 100) # starting point, ending point, level of detail
# show the data
scatter(distances, xLength)
scatter(distances, yLength)
# draw the line
plt.plot(xp, polynomial1(xp), c='g')
plt.plot(xp, polynomial2(xp), c='r')
plt.show()
```
